```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
x = np.arange(-50, 50, 1)
y = np.arange(-50, 50, 1)
x, y = np.meshgrid(x, y)
z2 = 4 - 2 * x - 3 * y
z1 = - 2 * (x + 50) - 3 * (y + 50)
ax.plot_surface(x, y, z1)
ax.plot_surface(x, y, z2)
plt.title('Параллельные плоскости')
plt.xlabel('X')
plt.ylabel('Y')
ax.plot_surface(x, y, z1, rstride = 2, cstride = 2, cmap = cm.inferno)
ax.plot_surface(x, y, z2, rstride = 2, cstride = 2, cmap = cm.inferno)
show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
xval = np.linspace(-10, 10, 100)
yval = np.linspace(-10, 10, 100)
f = lambda x, y: x ** 2 - y ** 2
x, y = np.meshgrid(xval, yval)
z = f(x, y)
surf = ax.plot_surface(x, y, z, rstride = 2, cstride = 2, cmap = cm.viridis)
plt.title('Гиперболический параболлойд')
plt.xlabel('X')
plt.ylabel('Y')



```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
xval = np.linspace(-10, 10, 100)
yval = np.linspace(-10, 10, 100)
f = lambda x, y: x ** 2 + y ** 2
x, y = np.meshgrid(xval, yval)
z = f(x, y)
surf = ax.plot_surface(x, y, z, rstride = 2, cstride = 2, cmap = cm.viridis)
plt.title('Эпиболический параболлойд')
plt.xlabel('X')
plt.ylabel('Y')
```
