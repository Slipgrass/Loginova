2. Задем программу для трехмерного графика, так как переменых 3


```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
x = np.arange(-1, 5, 0.5)
y = np.arange(-1, 5, 0.5)
plt.plot(x, y, x + 2 * y -1, 'red')
plt.plot(x, y, 0, 'blue')
plt.plot(x, y, 6 - 4 * x + 2.5 * y, 'green')
plt.plot(x, y, 0.4 * x - 1.4, 'orange')
plt.plot(x, y, (11 * x + 4 * y - 15) / 7, 'purple')
plt.xlabel('X')
plt.ylabel('Y')
show()
```

```python
A = np.array([[1, 2, -1], [3, -4, 0], [8, -5, 2], [2, -5, 0], [11, 4, -7]]) #записываем СЛАУ в виде матрицы
B = np.array([1, 0, 12, 7, 15])
np.linalg.lstsq(A, B, rcond=None) # поиск всевдорешения методом наименьших квадратов
```

```python
np.dot(A, [1.15601029, -0.889718  , -0.85029782]) - B #считаем норму невязки для каждого уравнения
```

```python
np.dot(A, [1.31925182, -0.21367343, -0.1620986]) #проверяем, насколько правильно метод ищет решение, подставив в него результаты
```

```python
def Q(x, y, z):
  return((np.linalg.norm(np.dot(A, [x, y, z]) - B))**2) #c помощью пользовательской функции считаем значения нормы невязки для СЛАУ

Q(1.31925182, -0.21367343, -0.1620986)

```

Ответ: 

Псевдорешением системы заданной СЛАУ являются значения [ 1.31925182, -0.21367343, -0.1620986]
