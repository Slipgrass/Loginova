```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-5, 5, 100)
r = int(input())
y1 = np.sqrt(r**2-x**2) 
y2 = -np.sqrt(r**2-x**2)
plt.plot(x, y1)
plt.plot(x, y2)
plt.title('График окружности')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show
```

```python
x = np.linspace(-5, 5, 100)
a = 5
b = 2
y1 = np.sqrt(b**2-(b**2*x**2/a**2)) 
y2 = -np.sqrt(b**2-(b**2*x**2/a**2))
plt.plot(x, y1)
plt.plot(x, y2)
plt.title('График')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show
```

```python
x1 = np.linspace(-10, 0, 100)
x2 = np.linspace(0, 10, 100)
y1 = 1/x1 
y2 = 1/x2
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.title('График гиперболы')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show
```
