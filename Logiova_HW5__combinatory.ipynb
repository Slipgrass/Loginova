```python
import numpy as np
import itertools
```

```python
for p in itertools.permutations("45678", 4):
    print(''.join(str(x) for x in p))   #вычисляется количество перестановок - размещений (А), учитывая
```

```python
for p in itertools.combinations("45678", 4):
    print(''.join(p))   #вычисляется количество размещений(без учета расположения объектов (С))
```

```python
for p in itertools.product("4567",repeat=4):
    print(''.join(p))   
```
