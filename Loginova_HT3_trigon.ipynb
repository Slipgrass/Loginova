```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 3*np.pi, 201)
plt.plot(x, np.cos(x))
plt.plot(x, -0.5*np.cos(x))
plt.plot(x, np.cos(x+1))
plt.plot(x, np.cos(x)-1)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
